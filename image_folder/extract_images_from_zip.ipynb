```python
import tarfile

# Extract the tar.gz file
with tarfile.open('cifar-10-python.tar.gz', 'r:gz') as tar:
    tar.extractall(path='cifar-10-batches-py')  # extract to a folder
```

```python
# Function to unpickle CIFAR-10 data files
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
# Load a specific batch file for train (batch 1) and test data
train_data_dict = unpickle('C:\\Users\\juhee\\OneDrive\\Documents\\Carnegie Mellon\\Fall 2025\\Deep Learning\\HWs\\hw5_student_starter_code\\image_folder\\cifar-10-batches-py\\cifar-10-batches-py\\data_batch_1')
test_data_dict = unpickle('C:\\Users\\juhee\\OneDrive\\Documents\\Carnegie Mellon\\Fall 2025\\Deep Learning\\HWs\\hw5_student_starter_code\\image_folder\\cifar-10-batches-py\\cifar-10-batches-py\\test_batch')
```

```python
print("Keys in train data dict:", train_data_dict.keys())
print("Length of train data:", len(train_data_dict[b'data']))
print("Length of test data:", len(test_data_dict[b'data']))

print("Unique labels in train data:", set(train_data_dict[b'labels']))
```

```python
# Create dummy train and test folders with extracted images

#split the data into train and test sets and save images accordingly
import os
import numpy as np
from PIL import Image
# Create directories
train_dir = 'cifar10_train'
test_dir = 'cifar10_test'
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Save training images into format accepted by ImageFolder
for i in range(len(train_data_dict[b'data'])):
    img_array = train_data_dict[b'data'][i].reshape(3, 32, 32).transpose(1, 2, 0)
    img = Image.fromarray(img_array)
    label = train_data_dict[b'labels'][i]
    label_dir = os.path.join(train_dir, str(label))
    os.makedirs(label_dir, exist_ok=True)
    img.save(os.path.join(label_dir, f'train_img_{i}.png'))

# Save test images into format accepted by ImageFolder
for i in range(len(test_data_dict[b'data'])):
    img_array = test_data_dict[b'data'][i].reshape(3, 32, 32).transpose(1, 2, 0)
    img = Image.fromarray(img_array)
    label = test_data_dict[b'labels'][i]
    label_dir = os.path.join(test_dir, str(label))
    os.makedirs(label_dir, exist_ok=True)
    img.save(os.path.join(label_dir, f'test_img_{i}.png'))
```
